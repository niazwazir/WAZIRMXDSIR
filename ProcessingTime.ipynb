```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *

psnr_scale_4 = [    25.53,   25.49,     25.38,   25.33,  25.93,   26.18,   26.12,     25.93,   26.22,          26.33]
no_of_parameters = [7,       100,      20,      0.002,   0.9,      5,      0.05,      19,      0.007,          0.005]
Name_Of_Models = ["A1",     "A2",   "A3",  "A4", "A5",  "A6",  "A7",  "A8",  "A9",     "A10"]
for k in range(10):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_4[k]
    if loc_x == 0.005 and loc_y == 26.33:
        plt.semilogx(loc_x,loc_y,color='b',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12,color='b')
    else:
        plt.semilogx(loc_x,loc_y,color='r',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12)
    
plt.ylim([25,26.4])
plt.xlim(100,0.0018)
#plt.xlim(10,1000)
plt.ylabel('PSNR(dB)',fontsize=16)
plt.xlabel('slower  <--------       running time(s)        ---------->        faster',fontsize=12)
plt.grid(which="both")
plt.show()
```
