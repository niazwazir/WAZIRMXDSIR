```python
DATADIR = "SRCNN-Keras/yang91"
DATADIR2 = "SRCNN-Keras/yang91"
#DATADIR2 = "DIV2K_train_HR"
```

```python
import os
import numpy as np
import cv2
```

```python
scale = 2
req_size = 20
inp_size = int(req_size/scale)
stride = 15

X=[]
labels=[]

for filera in os.listdir(DATADIR):
  readpath = os.path.join(DATADIR,filera)
  img = cv2.imread(readpath)
  img = np.asarray(img)
  shapes = img.shape
  for i in range(0, shapes[0]-req_size+1, stride):
    for j in range(0, shapes[1]-req_size+1,stride):
      subimage_HR = img[i:i+req_size, j:j+req_size]
      #cv2_imshow(subimage_HR)
      subimage_LR = cv2.resize(subimage_HR,(inp_size,inp_size))
      X.append(subimage_LR)
      labels.append(subimage_HR)

counta=0
for filera in os.listdir(DATADIR2):
  counta=counta+1
  if(counta==300):
    break
  readpath = os.path.join(DATADIR2,filera)
  img = cv2.imread(readpath)
  img = np.asarray(img)
  shapes = img.shape
  for i in range(0, shapes[0]-req_size+1, stride):
    for j in range(0, shapes[1]-req_size+1,stride):
      subimage_HR = img[i:i+req_size, j:j+req_size]
      #cv2_imshow(subimage_HR)
      subimage_LR = cv2.resize(subimage_HR,(inp_size,inp_size))
      X.append(subimage_LR)
      labels.append(subimage_HR)
```

```python
X = np.asarray(X)
labels = np.asarray(labels)
print(labels.shape)
print(X.shape)
```

```python
import tensorflow as tf
import keras
import numpy as np
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Conv2D, SeparableConv2D, Input,Conv2DTranspose, Activation
from tensorflow.keras.layers import LeakyReLU
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.layers import LeakyReLU
```

```python
alpha = 0.85
def basic_mse(y_true, y_pred):
  pred = tf.image.sobel_edges(y_pred)
  true = tf.image.sobel_edges(y_true)
  lossmse = tf.reduce_mean(tf.square(y_pred-y_true))
  lossedge = tf.reduce_mean(tf.square(pred-true))
  return alpha*lossmse + (1-alpha)*lossedge
```

```python
inputlayer = tf.keras.layers.Input(shape=(None,None,3))
#FIRST FEATURE EXTRACTION LAYER
C1 = Conv2D(64, (3,3), strides=(1,1), padding="same",name='CONV1')(inputlayer)
#A1 = tf.keras.layers.LeakyReLU(alpha=0.2)(C1)
A1 = LeakyReLU(alpha=0.2)(C1)

#FIRST RESNET BLOCK
C2 = Conv2D(64, (3,3),padding="same",name='CONV2')(A1)
#A2 = tf.keras.layers.LeakyReLU(alpha=0.3,name='A2')(C2)
A2 = LeakyReLU(alpha=0.2)(C2)
C3 = Conv2D(64, (3,3),padding="same",name='CONV3')(A2)
#A3 = Activation('relu', name='act3')(C3)
A3 = LeakyReLU(alpha=0.2)(C3)
Sum1  = tf.keras.layers.add([A1,A3])

#SECOND RESNET BLOCK
C4 = SeparableConv2D(64, (3,3),padding="same",name='CONV4')(Sum1)
#A4 = Activation('relu', name='act4')(C4)
A4 = LeakyReLU(alpha=0.2)(C4)

C5 = SeparableConv2D(64, (3,3),padding="same",name='CONV5')(A4)
#A5 = Activation('relu', name='act5')(C5)
A5 = LeakyReLU(alpha=0.2)(C5)
Sum2  = tf.keras.layers.add([A1,A5])

#SHRINKING LAYER
C6 = SeparableConv2D(64, (1,1),padding="same",name='CONV6')(Sum2)
#A6 = Activation('relu', name='act6')(C6)
A6 = LeakyReLU(alpha=0.2)(C6)

#DECONVOLUTION LAYER
y = Conv2DTranspose(3, (9,9), strides = (scale,scale), padding="same",activation=None,name='UPSCALING')(A6)

#EXPANDING LAYER 
C7 = SeparableConv2D(64, (1,1),padding="same",name='CONV7')(y)
#A7 = Activation('relu', name='act7')(C7)
A7 = LeakyReLU(alpha=0.2)(C7)
#XCEPTION BLOCK 1
X1 = SeparableConv2D(64, (3,3),padding="same",name='CONVX1')(A7)
#AX1 = Activation('relu', name='actx1')(X1)
AX1 = LeakyReLU(alpha=0.2)(X1)

X2 = SeparableConv2D(64, (5,5),padding="same",name='CONVX2')(A7)
#AX2 = Activation('relu', name='actx2')(X2)
AX2 = LeakyReLU(alpha=0.2)(X2)

X3 = SeparableConv2D(64, (7,7),padding="same",name='CONVX3')(A7)
#AX3 = Activation('relu', name='actx3')(X3)
AX3 = LeakyReLU(alpha=0.2)(X3)

Sum3  = tf.keras.layers.add([AX1,AX2, AX3])

C8 = SeparableConv2D(3, (3,3),padding="same",name='CONV8')(Sum3)
#A8 = Activation('relu', name='act8')(C8)
A8 = LeakyReLU(alpha=0.2)(C8)

#XCEPTION BLOCK 2
X4 = SeparableConv2D(64, (3,3),padding="same",name='CONVX4')(A8)
#AX4 = Activation('relu', name='actx4')(X4)
AX4 = LeakyReLU(alpha=0.2)(X4)

X5 = SeparableConv2D(64, (5,5),padding="same",name='CONVX5')(A8)
#AX5 = Activation('relu', name='actx5')(X5)
AX5 = LeakyReLU(alpha=0.2)(X5)

X6 = SeparableConv2D(64, (7,7),padding="same",name='CONVX6')(A8)
#AX6 = Activation('relu', name='actx6')(X6)
AX6 = LeakyReLU(alpha=0.2)(X6)

Sum4  = tf.keras.layers.add([AX4,AX5, AX6])

C9 = SeparableConv2D(3, (3,3),padding="same",name='CONV9')(Sum4)
#A9 = Activation('relu', name='act9')(C9)
A9 = LeakyReLU(alpha=0.2)(C9)

#XCEPTION BLOCK 3
X7 = SeparableConv2D(64, (3,3),padding="same",name='CONVX7')(A9)
#AX7 = Activation('relu', name='actx7')(X7)
AX7 = LeakyReLU(alpha=0.2)(X7)

X8 = SeparableConv2D(64, (5,5),padding="same",name='CONVX8')(A9)
#AX8 = Activation('relu', name='actx8')(X8)
AX8 = LeakyReLU(alpha=0.2)(X8)

X9 = SeparableConv2D(64, (7,7),padding="same",name='CONVX9')(A9)
#AX9 = Activation('relu', name='actx9')(X9)
AX9 = LeakyReLU(alpha=0.2)(X9)


Sum5  = tf.keras.layers.add([AX7,AX8, AX9])

C10 = Conv2D(3, (3,3),padding="same",name='CONV10')(Sum5)
#A10 = Activation('relu', name='act10')(C10)
A10 = LeakyReLU(alpha=0.2)(C10)


model = Model(inputs=inputlayer , outputs = A10)
model.summary()
```

```python
tf.keras.utils.plot_model(model, 'my_first_model.png')
```

```python
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam, SGD
```

```python
model.compile(loss='mse',
              optimizer='adam',
              metrics=["accuracy"])
```

```python
checkpoint = ModelCheckpoint('best_model_improved.h5',  # model filename
                             monitor='val_loss', # quantity to monitor
                             verbose=0, # verbosity - 0 or 1
                             save_best_only= True, # The latest best model will not be overwritten
                             mode='auto') # The decision to overwrite model is made 
                                          # automatically depending on the quantity to monitor
```

```python
history = model.fit(X,labels,batch_size=32,shuffle=True,validation_split=0.01 ,epochs=5,callbacks=[checkpoint],
                    verbose=1 )
```
